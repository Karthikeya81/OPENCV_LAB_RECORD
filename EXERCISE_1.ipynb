Load and Display a Video File

```python
# Import the 'files' module for uploading our video
from google.colab import files
# Import the clear_output tool so we can refresh the output screen frame-by-frame
from IPython.display import clear_output
# Print a helpful message asking for a video upload
print("Please upload a short video file (.mp4, .avi, etc.):")
# Prompt the user to upload a video file
uploaded_video = files.upload()
# Import OpenCV
import cv2
# Import the Colab-safe image display function
from google.colab.patches import cv2_imshow
# Get the filename of the uploaded video
video_filename = next(iter(uploaded_video))
# Create a VideoCapture object to load and read the video file
video_capture = cv2.VideoCapture(video_filename)
# Set a counter starting at zero to track how many frames we have shown
frame_count = 0
# Create a loop to read and display only the first 30 frames
while frame_count < 30:
    # Read the next frame; success is True if read correctly, frame holds the image data
    success, frame = video_capture.read()
    # Check if we failed to read a frame (for example, if the video ended)
    if not success:
        # Stop the loop immediately if there are no more frames
        break
    # Clear the previous frame's output so they don't create a giant vertical list
    clear_output(wait=True)
    # Display the current frame of the video
    cv2_imshow(frame)
    # Add 1 to our frame counter to keep track
    frame_count = frame_count + 1
# Release the video capture object to free up system memory
video_capture.release()
# Print a final success message to show the code finished smoothly
print("Finished displaying video frames!")
```
