Exercise - 6 (Thresholding)

```python
# --- Task 1: Import tools and upload ---
import cv2
from google.colab import files
from google.colab.patches import cv2_imshow

print("Upload an image (Try to find one with text, like a picture of a book page!):")
uploaded = files.upload()
filename = next(iter(uploaded))

# --- Task 2: Read the image in Grayscale ---
# We use the cv2.IMREAD_GRAYSCALE flag to convert it immediately as we load it!
# Thresholding ONLY works on grayscale images.
gray_image = cv2.imread(filename, cv2.IMREAD_GRAYSCALE)

# --- Task 3: Apply Thresholding Methods ---

# Method A: Simple/Global Thresholding
# We set the threshold value to 127. Everything above 127 becomes 255 (white).
# The function returns two things: the threshold value used (ret), and the new binary image (thresh_simple)
ret, thresh_simple = cv2.threshold(gray_image, 127, 255, cv2.THRESH_BINARY)

# Method B: Adaptive Thresholding
# Excellent for varying lighting conditions. It looks at an 11x11 grid and calculates the threshold for that specific spot.
thresh_adaptive = cv2.adaptiveThreshold(gray_image, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                                        cv2.THRESH_BINARY, 11, 2)

# Method C: Otsu's Thresholding
# We pass 0 as the threshold value because cv2.THRESH_OTSU tells OpenCV to calculate the best value automatically!
ret_otsu, thresh_otsu = cv2.threshold(gray_image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)


# --- Task 4: Display the Binary Images ---
print("\n--- Original Grayscale Image ---")
cv2_imshow(gray_image)

print("\n--- 1. Simple Thresholding (Fixed at 127) ---")
cv2_imshow(thresh_simple)

print("\n--- 2. Adaptive Thresholding (Adjusts for shadows) ---")
cv2_imshow(thresh_adaptive)

print(f"\n--- 3. Otsu's Thresholding (Auto-calculated optimal threshold: {ret_otsu}) ---")
cv2_imshow(thresh_otsu)
```
