Exercise - 5 (5. Image Filtering & Smoothing)

```python
# --- Task 1 & 2: Import OpenCV and Read Image ---
import cv2
from google.colab import files
from google.colab.patches import cv2_imshow

print("Please upload an image to apply filters to:")
uploaded = files.upload()
# Grab the filename and read the image
filename = next(iter(uploaded))
image = cv2.imread(filename)

# --- Task 3: Apply Different Filters ---

# A "kernel" is the size of the pixel grid the filter looks at.
# (5, 5) means a 5x5 square of pixels. Odd numbers are usually required!

# A. Averaging Filter
# Uses cv2.blur to simply average the pixels in the 5x5 grid.
avg_blur = cv2.blur(image, (5, 5))

# B. Gaussian Filter
# Uses cv2.GaussianBlur. The '0' at the end tells OpenCV to calculate the math variance automatically.
gaussian_blur = cv2.GaussianBlur(image, (5, 5), 0)

# C. Median Filter
# Uses cv2.medianBlur. It only needs a single integer for the kernel size (e.g., 5).
median_blur = cv2.medianBlur(image, 5)

# D. Bilateral Filter
# Uses cv2.bilateralFilter.
# Parameters are (image, neighborhood_diameter, sigma_color, sigma_space)
# 9 is the grid diameter, 75 determines how aggressively it blends colors, 75 blends spatial distances.
bilateral_blur = cv2.bilateralFilter(image, 9, 75, 75)


# --- Task 4: Display All Images ---
print("\n--- Original Image ---")
cv2_imshow(image)

print("\n--- 1. Averaging Filter (Simple Blur) ---")
cv2_imshow(avg_blur)

print("\n--- 2. Gaussian Filter (Natural Blur) ---")
cv2_imshow(gaussian_blur)

print("\n--- 3. Median Filter (Great for Specks/Noise) ---")
cv2_imshow(median_blur)

print("\n--- 4. Bilateral Filter (Smooths while keeping edges sharp!) ---")
cv2_imshow(bilateral_blur)
```
