Exercise - 3 (3. Image Processing Operations)

```python
# Import the tools we need
import cv2
from google.colab import files
from google.colab.patches import cv2_imshow

# Prompt you to upload an image from your computer
print("Please upload a test image (e.g., test_image.jpg):")
uploaded = files.upload()

# Grab the filename and read the image into a variable named 'image'
filename = next(iter(uploaded))
image = cv2.imread(filename)

print("Image successfully loaded! You are ready for the next tasks.")
```

```python
# --- 1. Resize using scaling factors ---
print("Task 1: Resizing (1.5x larger)")

# fx is the scale factor for the X-axis (width). 1.5 means 150%
# fy is the scale factor for the Y-axis (height). 1.5 means 150%
# interpolation=cv2.INTER_LINEAR is the mathematical method OpenCV uses to guess what the new pixels should look like
scaled_image = cv2.resize(image, None, fx=1.5, fy=1.5, interpolation=cv2.INTER_LINEAR)

# Display the result
cv2_imshow(scaled_image)
```

```python
# --- 2. Rotate an image ---
print("Task 2: Rotating by 45 degrees")

# First, get the height (h) and width (w) of the original image
(h, w) = image.shape[:2]

# Calculate the exact center coordinates (width divided by 2, height divided by 2)
# We use '//' for integer division because pixel coordinates must be whole numbers
center = (w // 2, h // 2)

# Create a rotation matrix using cv2.getRotationMatrix2D(center, angle, scale)
# We will rotate 45 degrees counter-clockwise, and keep the scale at 1.0 (original size)
rotation_matrix = cv2.getRotationMatrix2D(center, 45, 1.0)

# Apply the rotation matrix to the image using cv2.warpAffine
# We also pass in the original (width, height) to tell it how big the new canvas should be
rotated_image = cv2.warpAffine(image, rotation_matrix, (w, h))

# Display the result
cv2_imshow(rotated_image)
```

```python
# --- 3. Crop a specific portion ---
print("Task 3: Cropping a section")

# The syntax is image[startY:endY, startX:endX]
# Let's crop starting at Y=50 down to Y=200, and X=50 across to X=300
cropped_image = image[50:200, 50:300]

# Display the cropped piece
cv2_imshow(cropped_image)
```

```python
# --- 4. Flip the image ---
print("Task 4: Flipping Horizontally and Vertically")

# Flip horizontally (around the Y-axis) using flip code 1
horizontal_flip = cv2.flip(image, 1)
print("Horizontal Flip (Mirror):")
cv2_imshow(horizontal_flip)

# Flip vertically (around the X-axis) using flip code 0
vertical_flip = cv2.flip(image, 0)
print("Vertical Flip (Upside Down):")
cv2_imshow(vertical_flip)
```

```python
# --- 5. Adjust Brightness and Contrast ---
print("Task 5: High Contrast and High Brightness")

# Let's increase the contrast by 50% (alpha = 1.5)
# Let's add 50 units of brightness to every pixel (beta = 50)
adjusted_image = cv2.convertScaleAbs(image, alpha=1.5, beta=50)

# Display the newly lit image
cv2_imshow(adjusted_image)
```
