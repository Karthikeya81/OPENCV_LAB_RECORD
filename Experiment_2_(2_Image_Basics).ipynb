Experiment - 2 (2. Image Basics)

```python
# Import the tools we need
import cv2
from google.colab import files
from google.colab.patches import cv2_imshow

# Prompt you to upload an image from your computer
print("Please upload a test image (e.g., test_image.jpg):")
uploaded = files.upload()

# Grab the filename and read the image into a variable named 'image'
filename = next(iter(uploaded))
image = cv2.imread(filename)

print("Image successfully loaded! You are ready for the next tasks.")
```

Display Image Properties


```python
# 1. Image Properties
print("--- Image Properties ---")

# .shape returns a tuple of (Height, Width, Channels)
# Channels represent colors (e.g., 3 means Red, Green, Blue)
print("Shape (Height, Width, Channels):", image.shape)

# .size returns the total number of pixels/elements in the matrix
print("Total number of pixels/elements:", image.size)

# .dtype returns the image data type (usually uint8, meaning 8-bit unsigned integers from 0 to 255)
print("Data Type:", image.dtype)
```

Color Space Conversion (RGB, Grayscale, HSV)

```python
# 2. Color Space Conversions

# Convert the default BGR image to Grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
print("Grayscale Image:")
cv2_imshow(gray_image)

# Convert the BGR image to HSV (Hue, Saturation, Value)
hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
print("HSV Image:")
cv2_imshow(hsv_image)
```

```python
# 3. Split and Merge Channels

# Use cv2.split to separate the BGR image into 3 individual variables
b, g, r = cv2.split(image)

# Display the Blue channel.
# It will look like a grayscale image, where brighter white pixels mean "more blue" in the original image.
print("Blue Channel Isolated:")
cv2_imshow(b)

# Merge the variables back into a single color image using cv2.merge
merged_image = cv2.merge((b, g, r))
print("Re-Merged Image:")
cv2_imshow(merged_image)
```

```python
# 4. Extract Region of Interest (ROI)

# Let's crop a 150x150 square from the top-left corner of the image.
# We slice the Y-axis (height) from 0 to 150, and the X-axis (width) from 0 to 150.
roi = image[0:150, 0:150]

print("Region of Interest (Top-Left 150x150 pixels):")
cv2_imshow(roi)
```

```python
# 5. Change Resolution and Aspect Ratio


resized_absolute = cv2.resize(image, (250, 250))
print("Resized to exactly 250x250 pixels (Aspect ratio is ignored, might look squished):")
cv2_imshow(resized_absolute)

# --- Change Aspect Ratio / Scale ---
# Instead of absolute pixels, we use scale factors (fx and fy).
# fx=0.5 and fy=0.5 means we shrink both the width and height to 50% of the original size.
resized_scaled = cv2.resize(image, None, fx=0.5, fy=0.5, interpolation=cv2.INTER_LINEAR)
print("Scaled down to 50% (Aspect ratio is perfectly maintained):")
cv2_imshow(resized_scaled)
```
