Exercise - 4 (4. Drawing Functions)

```python
# Import OpenCV
import cv2
# Import NumPy to help us create the blank matrix (canvas)
import numpy as np
# Import the special Colab-safe image display function
from google.colab.patches import cv2_imshow

# Create a 500x500 canvas with 3 color channels (Blue, Green, Red)
# np.zeros creates a matrix filled with the number 0 (which means black)
# dtype="uint8" ensures the matrix uses the standard 8-bit image format
canvas = np.zeros((500, 500, 3), dtype="uint8")

print("Task 5: Blank Canvas Created!")
# Display the black canvas
cv2_imshow(canvas)
```

```python
# Draw a green diagonal line across the canvas
# The syntax is: cv2.line(image, start_point, end_point, color, thickness)
# Start at the top-left corner (0, 0)
# End at the bottom-right corner (500, 500)
# Color is BGR, so Green is (0, 255, 0)
# Thickness is set to 5 pixels
cv2.line(canvas, (0, 0), (500, 500), (0, 255, 0), 5)

print("Task 1: Line Drawn!")
# Display the updated canvas
cv2_imshow(canvas)
```

```python
# --- Draw a Rectangle ---
# Syntax: cv2.rectangle(image, top_left_corner, bottom_right_corner, color, thickness)
# Let's draw a blue rectangle. Blue in BGR is (255, 0, 0)
# A thickness of -1 tells OpenCV to fill the shape completely!
cv2.rectangle(canvas, (50, 50), (200, 200), (255, 0, 0), -1)

# --- Draw a Circle ---
# Syntax: cv2.circle(image, center_point, radius, color, thickness)
# Let's draw a red circle. Red in BGR is (0, 0, 255)
# Center it at (350, 150) with a radius of 75 pixels, and an outline thickness of 3 pixels
cv2.circle(canvas, (350, 150), 75, (0, 0, 255), 3)

print("Tasks 2 & 3: Rectangle and Circle Drawn!")
# Display the updated canvas showing the line, rectangle, and circle
cv2_imshow(canvas)
```

```python
# Define the font we want to use (built-in OpenCV font)
font = cv2.FONT_HERSHEY_SIMPLEX

# Write the text on the canvas
# Syntax: cv2.putText(image, text, bottom_left_corner, font, font_scale, color, thickness)
# Let's write "OpenCV is Fun!" in white: (255, 255, 255)
cv2.putText(canvas, "OpenCV is Fun!", (90, 450), font, 1.2, (255, 255, 255), 2)

print("Task 4: Text Added!")
# Display the final, completed canvas!
cv2_imshow(canvas)
```
